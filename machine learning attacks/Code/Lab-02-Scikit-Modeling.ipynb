# Write Accuracy Function

```python
def get_accuracy(cfile: str, rfile: str, p: float) -> float:
    crpData = ChallengeResponseSet(cfile, rfile)
    x,y = crpData.train_test_split(p)
    x_train = crpData.train.challenges
    y_train = crpData.train.responses
    x_test = crpData.test.challenges
    y_test = crpData.test.responses
    classifier = MLPClassifier().fit(x_train, y_train)
    accuracy =  classifier.score(x_test, y_test)
    return accuracy
```
