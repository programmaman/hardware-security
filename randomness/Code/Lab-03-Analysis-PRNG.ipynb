# Test and get P-Values

Instantiate a `TRNGtester` object to load the random data. Then, call the `run_nist_tests()` method from this object to generate the _list of list_ of p-values. Each sub-list correspond to each test. Some test may have multiple p-values -- only use the _minimum_ from each test.

Repeat this for each dataset (i.e., 1M and 10M bits from both the RO/LFSR TRNG and the GARO TRNG)

```python
test1 = TRNGtester("../Data/NUMPY_1M.bin")
test2 = TRNGtester("../Data/OS_1M.bin")
test3 = TRNGtester("../Data/NUMPY_10M.bin")
test4 = TRNGtester("../Data/OS_10M.bin")
results_numpy_1M = test1.run_nist_tests()
results_os_urandom_1M = test2.run_nist_tests()
results_numpy_10M = test3.run_nist_tests()
results_os_urandom_10M = test4.run_nist_tests()
print("✔️ OK to continue!")
# results
# for i in results:
#     print(min(i))
```

# Stats

Calculate: how many tests passed (p-value ≥ 0.01) or failed (p-value < 0.01)?

Repeat this for each dataset (i.e., 1M and 10M bits from both the RO/LFSR TRNG and the GARO TRNG)

```python
count_pass = 0
count_fail = 0
NP_1M_list = []
for i in results_numpy_1M:
    NP_1M_list.append(np.min(i))
    print(min(i))
    if min(i) < 0.01:
        count_fail += 1
    elif min(i) >= 0.01:
        count_pass += 1

print("Tests failed:", count_fail)
print("Tests passed:", count_pass)
```

# Plots

Plot and save (as PDF) a vertical bar graph for the NIST test. Make sure to format the plots with adequate size, title, axis labels, ... (see the Tutorial-01-Python).

To generate vertical bar graphs, use `plt.bar()`. Read more: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

Make sure to annotate each bar indicating if the test `pass` or `fail`, use `plt..annotate()`. Read more: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html

```python
results = NP_1M_list

x_axis = []
pass_fail = []


for i in range(len(results)):
    x_axis.append("T" + str(i))
    if results[i] < 0.01:
        pass_fail.append("X")
    elif results [i] > 0.01:
        pass_fail.append("✔️")




# for i in range(len(results)):
#     x_axis.append("Test" + str(i))


# figure properties
plt.figure(figsize=(10,5), constrained_layout=True)


# plots the data
plt.bar(x_axis, results)
plt.annotate(pass_fail[0], (0,0.5))
plt.annotate(pass_fail[1], (1,0.5))
plt.annotate(pass_fail[2], (2,0.5))
plt.annotate(pass_fail[3], (3,0.5))
plt.annotate(pass_fail[4], (4,0.5))
plt.annotate(pass_fail[5], (5,0.5))
plt.annotate(pass_fail[6], (6,0.5))
plt.annotate(pass_fail[7], (7,0.5))
plt.annotate(pass_fail[8], (8,0.5))
plt.annotate(pass_fail[9], (9,0.5))
plt.annotate(pass_fail[10], (10,0.5))
plt.annotate(pass_fail[11], (11,0.5))
plt.annotate(pass_fail[12], (12,0.5))
plt.annotate(pass_fail[13], (13,0.5))
plt.annotate(pass_fail[14], (14,0.5))


# # format your plots
plt.title("NP 1 Million Bits Test p-values Results")            # adds title
plt.xlabel("NIST Test")          # adds x-axis label
plt.ylabel("p-score")          # adds y-axis label
# plt.xlim(0, 3)               # sets x-axis min/max values
# plt.ylim(0, 1.01)             # sets y-axis min/max values


# saves the plot
plt.savefig("../Figures/NP_1M.pdf") # PDF recommended



# show the plot on your screen
plt.show()
```
